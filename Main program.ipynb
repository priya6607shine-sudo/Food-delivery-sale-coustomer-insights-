```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class DeliveryAnalytics:
    def __init__(self, csv_file=None):
        # If no file is provided, generate sample data for demonstration
        if csv_file:
            self.df = pd.read_csv(csv_file)
        else:
            self.df = self._generate_sample_data()

    def _generate_sample_data(self):
        data = {
            'order_id': range(1, 101),
            'item': ['Burger', 'Pizza', 'Sushi', 'Pasta', 'Salad'] * 20,
            'amount': [15, 25, 30, 20, 12] * 20,
            'delivery_time_mins': [25, 40, 35, 30, 20] * 20
        }
        return pd.DataFrame(data)

    def get_summary(self):
        print("--- Performance Summary ---")
        print(f"Total Orders: {len(self.df)}")
        print(f"Total Revenue: ${self.df['amount'].sum():.2f}")
        print(f"Average Order Value: ${self.df['amount'].mean():.2f}")
        print("-" * 27)

    def plot_top_items(self):
        plt.figure(figsize=(10, 5))
        item_sales = self.df.groupby('item')['amount'].sum().sort_values()
        item_sales.plot(kind='barh', color='teal')
        plt.title('Revenue by Menu Item')
        plt.xlabel('Revenue ($)')
        plt.show()

    def analyze_delivery_efficiency(self):
        # Correlation between order value and delivery time
        sns.scatterplot(data=self.df, x='amount', y='delivery_time_mins', hue='item')
        plt.title('Order Value vs. Delivery Time')
        plt.show()

# --- Execution ---
# Initialize the analyzer
analyzer = DeliveryAnalytics()

# Run reports
analyzer.get_summary()
analyzer.plot_top_items()
analyzer.analyze_delivery_efficiency()
```
